```python
import matplotlib.pyplot as plt
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Category wise Sales')
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title('Category wise Profit')
plt.show()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values().head(10)
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title('Region wise Profit')
plt.show()
```

```python
plt.scatter(df['Discount'], df['Profit'])
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

###  Key Insights

- Technology category highest sales & profit
- Furniture category low profit
- Tables & Bookcases major loss products
- High discount orders mostly lead to loss
- West region more profitable
